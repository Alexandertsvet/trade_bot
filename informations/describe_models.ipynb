Модель FinancialInstrument

# Основные идентификаторы
* name: Название компании или фонда.
* ticker: Краткий биржевой код (например, SBER).
* class_code: Код класса торгов (например, TQBR для акций на Мосбирже).
* figi: Международный уникальный идентификатор финансового инструмента.
* uid: Внутренний уникальный идентификатор в системе Т-Банка.
# Параметры торговли
* type: Тип ценной бумаги (share, bond, etf, currency).
* min_price_increment: Минимальный шаг изменения цены (квант).
* scale: Количество знаков после запятой для цены.
* lot: Размер одного лота (минимальное количество бумаг в одной сделке).
* trading_status: Текущий статус торгов (например, доступны торги или сессия закрыта).
# Ограничения и флаги (True/False)
* api_trade_available_flag: Доступна ли торговля этим инструментом через API.
* currency: Валюта, в которой номинирован инструмент (rub, usd).
* exchange: Биржа, где торгуется актив (MOEX, SPBEX).
* buy_available_flag: Разрешена ли покупка в данный момент.
* sell_available_flag: Разрешена ли продажа в данный момент.
* short_enabled_flag: Разрешены ли маржинальные сделки Шорт (продажа без покрытия).
# Риск-параметры
* klong: Коэффициент ставки риска для длинных позиций (Лонг).
* kshort: Коэффициент ставки риска для коротких позиций (Шорт).
